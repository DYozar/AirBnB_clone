## 1. Cos'è un'eccezione?

Quando esegui un programma Python, può succedere che durante l'esecuzione capiti qualcosa di **anomalo**: provi a dividere per zero, a leggere un file che non esiste, a convertire in numero la parola `"ciao"`...

In questi casi Python non sa come continuare, quindi **solleva un'eccezione** (in inglese *raise an exception*). Un'eccezione è un oggetto speciale che dice: *"si è verificato un problema di tipo X"*.

Se nessuno **gestisce** quell'eccezione, il programma si **interrompe** e mostra un messaggio di errore chiamato **traceback**.

Vediamo subito cosa succede quando un'eccezione **non** viene gestita.


````python
# Proviamo a dividere un numero per zero.
# Questa cella darà ERRORE di proposito: serve a vedere com'è fatto un traceback.

numero = 10
risultato = numero / 0   # <-- divisione per zero: impossibile!
print(risultato)         # questa riga NON verrà mai eseguita
````

Hai ottenuto qualcosa del genere:

```
ZeroDivisionError: division by zero
```

Leggiamo il traceback **dal basso verso l'alto**:

- L'ultima riga dice il **tipo** di eccezione (`ZeroDivisionError`) e un **messaggio** (`division by zero`).
- Le righe sopra indicano **dove** nel codice è avvenuto l'errore.

Nota anche che `print(risultato)` **non** è stato eseguito: appena scatta l'eccezione, Python interrompe tutto.


## 2. Errori di sintassi ≠ eccezioni

È importante distinguere due cose diverse:

| | Errore di sintassi (`SyntaxError`) | Eccezione |
|---|---|---|
| **Quando avviene** | *Prima* di eseguire, quando Python legge il codice | *Durante* l'esecuzione |
| **Causa** | Hai scritto codice che non rispetta le regole del linguaggio | Il codice è valido ma succede qualcosa di anomalo |
| **Esempio** | `print("ciao"` (parentesi non chiusa) | `10 / 0` |
| **Si può gestire con try/except?** | ❌ No | ✅ Sì |

In altre parole: un errore di sintassi è come una frase scritta male grammaticalmente (non si può nemmeno leggere); un'eccezione è una frase corretta che però chiede una cosa impossibile.

Le **eccezioni** sono quelle che impareremo a gestire.


````python
# Esempio di codice SINTATTICAMENTE valido che però solleva eccezioni
# a seconda dei dati. La sintassi è perfetta, ma i dati creano problemi.

print(int("123"))    # OK: "123" si può convertire in numero -> 123
print(int("ciao"))   # ERRORE: "ciao" non è un numero -> ValueError
````

## 4. Il blocco `try` / `except`

Finora i nostri programmi si bloccavano. Ora impariamo a **gestire** le eccezioni così che il programma possa continuare o reagire in modo controllato.

Lo strumento base è il blocco `try` / `except`:

```python
try:
    # codice "rischioso" che potrebbe sollevare un'eccezione
except:
    # codice da eseguire SE è scattata un'eccezione
```

Come funziona:

1. Python esegue il codice dentro `try`.
2. Se **tutto va bene**, il blocco `except` viene **ignorato**.
3. Se scatta un'eccezione, Python **interrompe** subito il `try` e salta nel blocco `except`.
4. Dopo l'`except`, il programma **prosegue** normalmente: niente più crash!


````python
# Stessa divisione per zero di prima, ma stavolta la gestiamo.

try:
    numero = 10
    risultato = numero / 0
    print(risultato)            # NON viene eseguito
except:
    print("Attenzione: non puoi dividere per zero!")

print("Il programma continua tranquillamente. 🙂")
````

Nota la differenza con prima: il programma **non** si è bloccato. Ha eseguito il messaggio dell'`except` ed è andato avanti fino all'ultima `print`.


````python
# Esempio classico: chiedere un numero all'utente.
# input() restituisce sempre una STRINGA, e int() può fallire se l'utente
# scrive qualcosa che non è un numero.

try:
    testo = input("Scrivi un numero intero: ")
    valore = int(testo)
    print("Il doppio è:", valore * 2)
except:
    print("Quello che hai scritto non è un numero intero valido.")
````

## 5. Catturare eccezioni *specifiche*

Scrivere `except:` "nudo" cattura **qualunque** eccezione. Comodo, ma **pericoloso**: nasconde anche errori che non avevi previsto (per esempio un errore di battitura in un nome di variabile), rendendo difficile capire cosa sta succedendo davvero.

La buona pratica è specificare **quale** eccezione vuoi gestire:

```python
try:
    ...
except ValueError:
    # gestisco SOLO i ValueError
```

Così cattturi solo l'errore che ti aspetti; tutti gli altri continuano a comparire normalmente (ed è giusto che sia così, perché sono bug da sistemare).


````python
# Catturiamo SOLO il ValueError generato da int().

try:
    testo = input("Scrivi un numero intero: ")
    valore = int(testo)
    print("Hai scritto il numero:", valore)
except ValueError:
    print("Errore: questo non è un numero intero!")
````

## 6. Gestire più eccezioni

Spesso un pezzo di codice può fallire in **modi diversi**. Hai tre possibilità:

**A) Un `except` per ogni tipo** (reazioni diverse a errori diversi):

```python
try:
    ...
except ValueError:
    print("Valore non valido")
except ZeroDivisionError:
    print("Divisione per zero")
```

**B) Più tipi in un solo `except`** (stessa reazione), usando una tupla `( ... )`:

```python
try:
    ...
except (ValueError, ZeroDivisionError):
    print("C'è stato un problema con i numeri")
```

Python prova gli `except` **dall'alto verso il basso** e usa il **primo** che corrisponde al tipo dell'eccezione.


````python
# Esempio A: reazioni diverse per errori diversi.
# Programma che divide 100 per un numero inserito dall'utente.

try:
    n = int(input("Per quale numero vuoi dividere 100? "))
    print("Risultato:", 100 / n)
except ValueError:
    print("Devi inserire un numero, non del testo.")
except ZeroDivisionError:
    print("Non posso dividere per zero.")
````

````python
# Esempio B: stessa reazione per più errori, raggruppati in una tupla.

dati = ["10", "ciao", "0", "5"]

for elemento in dati:
    try:
        print("100 /", elemento, "=", 100 / int(elemento))
    except (ValueError, ZeroDivisionError):
        print("Impossibile usare il valore:", elemento)
````

## 7. La parola chiave `as`: leggere il dettaglio dell'errore

A volte vuoi sapere **esattamente** cosa è andato storto, non solo che è successo qualcosa. Con `as` catturi l'**oggetto eccezione** in una variabile e puoi stamparlo:

```python
try:
    ...
except ValueError as errore:
    print("Dettaglio:", errore)
```

L'oggetto eccezione, convertito in stringa, contiene il **messaggio** di Python (lo stesso che vedevi nel traceback).


````python
# Stampiamo il messaggio preciso dell'eccezione.

try:
    valore = int("quarantadue")
except ValueError as e:
    print("È scattata un'eccezione.")
    print("Tipo:", type(e).__name__)   # nome del tipo, es. 'ValueError'
    print("Messaggio:", e)             # il testo dell'errore
````

## 8. Il blocco `else`

Al `try`/`except` puoi aggiungere un blocco `else`. Il codice nell'`else` viene eseguito **solo se NON è scattata alcuna eccezione**:

```python
try:
    ...            # codice rischioso
except ...:
    ...            # eseguito SOLO se c'è un errore
else:
    ...            # eseguito SOLO se NON c'è stato alcun errore
```

A cosa serve? A separare chiaramente:
- il codice che **può fallire** (dentro `try`),
- il codice da eseguire **dopo, solo se è andato tutto bene** (dentro `else`).

In questo modo il `try` resta piccolo e contiene solo la riga davvero rischiosa.


````python
# Senza else metteremmo tutto dentro il try.
# Con else teniamo nel try SOLO la conversione (la parte che può fallire).

testo = "50"

try:
    numero = int(testo)          # parte rischiosa
except ValueError:
    print("Non è un numero valido.")
else:
    # qui siamo sicuri che 'numero' esiste ed è valido
    print("Conversione riuscita!")
    print("Il quadrato è:", numero ** 2)
````

## 9. Il blocco `finally`

Il blocco `finally` viene eseguito **SEMPRE**, sia che ci sia stata un'eccezione sia che non ci sia stata. È pensato per le operazioni di "pulizia" che vanno fatte in ogni caso: chiudere un file, chiudere una connessione, liberare risorse...

```python
try:
    ...
except ...:
    ...
finally:
    ...   # eseguito SEMPRE, comunque vada
```

> **Per chi viene dal C/C++:** è un po' come mettere il codice di cleanup in un punto garantito invece di doverlo ripetere prima di ogni `return`. In C++ è il ruolo dei distruttori (RAII); in Python lo fa `finally`.


````python
# Dimostriamo che finally viene eseguito in ENTRAMBI i casi.

def prova_divisione(a, b):
    try:
        print("Provo a calcolare", a, "/", b)
        risultato = a / b
        print("Risultato:", risultato)
    except ZeroDivisionError:
        print("Errore: divisione per zero!")
    finally:
        print("--> finally: questo viene SEMPRE eseguito.\n")

prova_divisione(10, 2)   # caso senza errore
prova_divisione(10, 0)   # caso con errore
````

Nota in particolare il secondo caso: anche se è scattata l'eccezione, il messaggio del `finally` è comparso lo stesso.

L'uso più comune di `finally` è con i **file**:

```python
file = open("dati.txt")
try:
    contenuto = file.read()
finally:
    file.close()   # il file viene chiuso anche se la lettura fallisce
```

> In pratica, per i file si usa spesso il costrutto `with open(...) as f:` che chiude il file automaticamente. Ma sotto al cofano l'idea è la stessa di `finally`.


## 10. Sollevare eccezioni con `raise`

Finora abbiamo **gestito** eccezioni sollevate da Python. Ma puoi anche **sollevarle tu** di proposito, con `raise`, quando vuoi segnalare che qualcosa non va.

```python
raise ValueError("messaggio che spiega il problema")
```

È utilissimo per **validare i dati** all'inizio di una funzione: se i dati sono sbagliati, sollevi subito un'eccezione chiara invece di lasciar proseguire un calcolo con valori assurdi.


````python
# Una funzione che rifiuta età negative sollevando un'eccezione.

def registra_eta(eta):
    if eta < 0:
        raise ValueError("L'età non può essere negativa: hai passato " + str(eta))
    print("Età registrata:", eta)

# Caso valido
registra_eta(25)

# Caso non valido: solleva ValueError (questa cella terminerà con errore)
registra_eta(-3)
````

Ovviamente, chi chiama la funzione può **gestire** l'eccezione che hai sollevato, esattamente come faresti con un errore di Python:


````python
def registra_eta(eta):
    if eta < 0:
        raise ValueError("L'età non può essere negativa.")
    print("Età registrata:", eta)

# Chi chiama decide come reagire all'eccezione.
for valore in [30, -5, 18]:
    try:
        registra_eta(valore)
    except ValueError as e:
        print("Valore rifiutato:", e)
````

## 11. Eccezioni personalizzate

Oltre alle eccezioni "standard" di Python, puoi crearne di **tue**, con un nome che descrive esattamente il tuo problema. Basta definire una classe che **eredita** da `Exception`:

```python
class SaldoInsufficienteError(Exception):
    pass
```

Poi la sollevi con `raise` come qualsiasi altra. Il vantaggio è la **leggibilità**: `SaldoInsufficienteError` dice molto più di un generico `ValueError`.

> Non serve conoscere a fondo le classi: per ora ti basta sapere che `class NomeError(Exception): pass` crea un nuovo tipo di eccezione pronto all'uso.


````python
# Definiamo un'eccezione personalizzata per un piccolo conto bancario.

class SaldoInsufficienteError(Exception):
    pass

def preleva(saldo, importo):
    if importo > saldo:
        raise SaldoInsufficienteError(
            "Saldo insufficiente: hai " + str(saldo) + "€, ne servono " + str(importo) + "€"
        )
    nuovo_saldo = saldo - importo
    print("Prelievo riuscito. Nuovo saldo:", nuovo_saldo, "€")
    return nuovo_saldo

# Proviamo a gestirla
try:
    preleva(saldo=100, importo=150)
except SaldoInsufficienteError as e:
    print("Operazione annullata.")
    print("Motivo:", e)
````

## 12. Buone pratiche (e cosa evitare)

✅ **Cattura eccezioni specifiche**, non `except:` generico.
✅ **Tieni piccolo il blocco `try`**: solo la riga che può fallire. Il resto va in `else`.
✅ **Scrivi messaggi chiari** quando usi `raise`.
✅ **Usa `finally`** (o `with`) per chiudere file e risorse.

❌ **Non "ingoiare" gli errori** con un `except` vuoto che non fa nulla:

```python
try:
    operazione_importante()
except:
    pass          # ❌ MALE: se c'è un bug, non lo scoprirai mai!
```

❌ **Non usare le eccezioni al posto dei normali controlli** quando un semplice `if` basta. Le eccezioni servono per situazioni *eccezionali*, non per il flusso normale.

❌ **Non catturare un'eccezione se non sai come reagire.** Meglio lasciarla salire fino a chi sa gestirla.
